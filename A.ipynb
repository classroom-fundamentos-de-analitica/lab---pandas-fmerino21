### Pregunta #12

```python
def format(df):
    aux_1 = [str(i) for i in df['_c5a']]
    aux_2 = [str(i) for i in df['_c5b']]
    aux = sorted(list(zip(aux_1, aux_2)))
    ans = []
    for i in aux:
        s = i[0] + ":" + str(i[1])
        ans.append(s)

    return ",".join(ans)
```

```python
ans = tbl2.groupby('_c0').apply(format).reset_index()
ans.rename(columns={0: '_c5'}, inplace=True)
ans
```
